### <a name='1'></a> Import bibliotek

````
import plotly
import plotly.graph_objects as go
import plotly.express as px
plotly.__version__
````

### <a name='2'></a> Wykres słupkowy

````
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4])
)
fig.show()
````

````
df = px.data.carshare()
df.head()
````

````
fig = go.Figure(
    data=go.Bar(
        x=df['centroid_lat'],            # Etykiety osi x - np. szerokość geograficzna
        y=df['car_hours'],               # Wartości osi y - liczba godzin dostępności samochodów
        text=df['centroid_lon'],         # Dodatkowy tekst przy słupkach (np. długość geograficzna)
        hovertext=df['peak_hour'],       # Tekst wyświetlany po najechaniu kursorem (np. godzina szczytu)
        marker_color=df['car_hours'],    # Kolorowanie słupków na podstawie liczby godzin
        marker_colorscale='Viridis',     # Skala kolorów
        marker_line_width=4,           # Szerokość linii obramowania słupków
        opacity = 0.6,
        marker_line_color='#666666'        # Kolor linii obramowania słupków
    )
)

fig.update_layout(
    title = "Wykres słupkowy - wypożyczalnia aut",
    xaxis_title = "Szerokość geograficzna (centroid_lat)",
    yaxis_title = "Liczba godzin (car_hours)",
    coloraxis_colorbar = dict(title = "Godziny użytkownika")
    )

fig.show()
````

````
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title={'text': 'Wykres słupkowy'})
)
fig.show()
````

````
fig = go.Figure(
    data=go.Bar(y=[2, 3, 1, 4]),
    layout=go.Layout(title_text='Wykres słupkowy')
)
fig.write_html('bar.html')
fig.show()
````

### <a name='3'></a> Wykres kołowy i pierścieniowy

````
from seaborn import load_dataset
df = load_dataset('diamonds')
df.head()
````

````
df.info()
````

````
dff = df.cut.value_counts()
dff = dff.reset_index()
dff
````

````
dff = df.cut.value_counts()
dff = dff.reset_index()
dff.columns = ['type', 'count']
dff
````

````
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count'])
)
fig.show()
````

````
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count'], hole=0.5),
    layout=go.Layout(title_text='Rozkład zmiennej cut', width=800)
)
fig.show()
````

### <a name='4'></a> Diagram Sankey'a

````
data = [go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019']),
                 link=dict(source=[0, 0, 1, 1], # indeks odpowiadający etykiecie (labels)
                          target=[2, 3, 2, 3],
                          value=[65, 12, 18, 5]))]

fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
````

````
data = [go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019',
                                   'Churn_2019', 'Nonchurn_2020', 'Churn_2020']),
                 link=dict(source=[0, 0, 1, 1, 2, 3], # indeks odpowiadający etykiecie (labels)
                          target=[2, 3, 2, 3, 4, 5],
                          value=[65, 10, 5, 20, 70, 30]))]

fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
````

### <a name='5'></a> Wykres świecowy

````
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df = df.reset_index()
df = df[df.Date > '2019-01-01']
df.head(3)
````

````
fig = go.Figure(data=go.Candlestick(x=df.Date, open=df.Open, high=df.High, low=df.Low, close=df.Close))
fig.show()
````
