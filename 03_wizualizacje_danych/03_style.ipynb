````
import numpy as np
import matplotlib.pyplot as plt
````

````
x = np.arange(-10, 10, 0.1)
x_sin = np.sin(x)

plt.plot(x, x_sin)
````

````
plt.style.use('ggplot')
plt.plot(x, x_sin)
````

````
plt.style.available
````

````
plt.style.use('bmh')
plt.plot(x, x_sin)
````

````
plt.style.use('dark_background')
plt.plot(x, x_sin)
````

````
plt.style.use('classic')
plt.plot(x, x_sin)
````
